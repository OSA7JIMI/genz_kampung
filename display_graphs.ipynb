```python
def favs(df, attribute, proportion, maximum):
    s = []
    for sub in df[attribute].unique():
        s.append({
            'sub': sub, 'count': len(df[df[attribute]==sub])
        })

    s_df = pd.DataFrame(s)
    s1 = s_df.sort_values('count',ascending=False)
    res = []

    for i in range(min(len(s1),maximum)):
        res.append(s1.iloc[i]['sub'])
    return res
```

# Survival Graphs (Topic Attention Span)

```python
p_stats = []
for u in p:
    au =  u + '.csv'
    try:
        top = pd.read_csv('Posters/Topics/' + au)
    except:
        continue
        
    shift = top['angle'][:-1]
    n_bins = []
    x = 1
    for i in range(len(shift)):
        if shift[i]>=45:
            n_bins.append(x)
            x = 1
        else:
            x+=1
            
    if len(n_bins) == 0:
        p_stats.append({
            'name': u, 'max': 99999, 'mean': 99999, 'first': 99999
        })
        continue
    
    p_stats.append({
        'name': u, 'max': max(n_bins), 'mean': sum(n_bins)/len(n_bins), 'first': n_bins[0]
    })

c_stats = []
for u in c:
    au =  u + '.csv'
    try:
        top = pd.read_csv('Commenters/Topics/' + au)
    except:
        continue
        
    shift = top['angle'][:-1]
    n_bins = []
    x = 1
    for i in range(len(shift)):
        if shift[i]>=45:
            n_bins.append(x)
            x = 1
        else:
            x+=1
    
    if len(n_bins) == 0:
        c_stats.append({
            'name': u, 'max': 99999, 'mean': 99999, 'first': 99999
        })
        continue
    
    c_stats.append({
        'name': u, 'max': max(n_bins), 'mean': sum(n_bins)/len(n_bins), 'first': n_bins[0]
    })
    
pstats = pd.DataFrame(p_stats)
cstats = pd.DataFrame(c_stats)
```

```python
gp = pstats[pstats['max']<99999]
np = len(pstats)
xp_max = npy.array([i for i in gp['max'].unique()])
xp_mean = npy.array([i for i in gp['mean'].unique()])
xp_first = npy.array([i for i in gp['first'].unique()])
xp_max.sort()
xp_mean.sort()
xp_first.sort()
yp_max = []
yp_mean = []
yp_first = []
```

```python
for x in xp_max:
    lp = len(pstats[pstats['max']>x])
    yp_max.append(lp/np)

for x in xp_mean:
    lp = len(pstats[pstats['mean']>x])
    yp_mean.append(lp/np)
    
for x in xp_first:
    lp = len(pstats[pstats['first']>x])
    yp_first.append(lp/np)
    
xp_max = npy.insert(xp_max, 0,0)
xp_mean = npy.insert(xp_mean, 0,0)
xp_first = npy.insert(xp_first, 0,0)

yp_max.insert(0,1)
yp_mean.insert(0,1)
yp_first.insert(0,1)
```

```python
gc = cstats[cstats['max']<99999]
nc = len(cstats)
xc_max = npy.array([i for i in gc['max'].unique()])
xc_mean = npy.array([i for i in gc['mean'].unique()])
xc_first = npy.array([i for i in gc['first'].unique()])
xc_max.sort()
xc_mean.sort()
xc_first.sort()
yc_max = []
yc_mean = []
yc_first = []
```

```python
for x in xc_max:
    lc = len(cstats[cstats['max']>x])
    yc_max.append(lc/nc)

for x in xc_mean:
    lc = len(cstats[cstats['mean']>x])
    yc_mean.append(lc/nc)
    
for x in xc_first:
    lc = len(cstats[cstats['first']>x])
    yc_first.append(lc/nc)
    
xc_max = npy.insert(xc_max, 0,0)
xc_mean = npy.insert(xc_mean, 0,0)
xc_first = npy.insert(xc_first, 0,0)

yc_max.insert(0,1)
yc_mean.insert(0,1)
yc_first.insert(0,1)
```

```python
plt.figure(figsize=(20,6))

plt.plot(xp_max, yp_max, marker='o', linestyle='-', color='b',label = 'Posters')
plt.plot(xc_max, yc_max, marker='o', linestyle='-', color='r',label = 'Commenters')

# plt.xlim(0, max(max(xp_max),max(xc_max))+1)
plt.xlim(0,140)

plt.xlabel('Number of bins')
plt.ylabel('% of users')
plt.title('Survival Graph (max)')

plt.legend()
plt.show()
```
